```python
# import pandas
import pandas as pd
# Import the datetime module from the datetime library.
from datetime import datetime
from datetime import timedelta
# Import the requests library.
import requests
# import socrata
from sodapy import Socrata
# import json
import json
# import numpy
import numpy as np
```

```python
# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("data.nashville.gov", None)

# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get("3wb6-xy3j", limit=3000)

# Convert to pandas DataFrame
df = pd.DataFrame.from_records(results)
```

```python
# check dataframe
df
```

```python
# select necessary columns
df= df[['permit_number', 'permit_subtype', 'date_entered', 'date_issued', 'business_name', 'zip', 'mapped_location']].copy() 
```

```python
df.head()
```

```python
# locate duplicate permit_number
df.loc[df.duplicated(subset = ["permit_number"])]
```

```python
# drop the duplicate permit_number rows
df.drop_duplicates(subset = ["permit_number"], inplace = True)
```

```python
# check duplicates were removed
df.loc[df.duplicated(subset = ["permit_number"])]
```

```python
# drop null rows
df = df.dropna()
df.head(10)
len(df)
```

```python
# drop offsite sales
df.drop(index=df[df['permit_subtype']=="OFFSALES"].index, axis=0, inplace=True)
df.drop(index=df[df['permit_subtype']=="WHOLESALES"].index, axis=0, inplace=True)
```

```python
# extract latitude and longitude
df['latitude'] = df['mapped_location'].apply(lambda x: x.get('latitude'))
df['longitude'] = df['mapped_location'].apply(lambda x: x.get('longitude'))
df.drop(columns = ['mapped_location'], inplace=True)
df.head()
```

```python
# create new column with timedelta of date issued and today
df['permit_duration'] = pd.Timestamp.now().normalize() - pd.to_datetime(df['date_issued']) 
df.head()
len(df)
```

```python
#df['date_entered', 'date_issued'] = pd.to_datetime(df, yearfirst=True)
df['date_entered'] = pd.to_datetime(df['date_entered'], infer_datetime_format=True)
df['date_issued'] = pd.to_datetime(df['date_issued'], infer_datetime_format=True)
df.head()
```
