```python
#Import dependencies
import pandas as pd
import numpy as np
```

```python
#Pull beer permit dataset into dataframe
beer_permit = pd.read_csv('Resources/Beer_Permit_Locations.csv', low_memory=False)
beer_permit.head()
```

```python
#Determine if there are missing values in dataframe
beer_permit.isnull().sum()
```

```python
#Drop Business Owner, Council District, and Census Tract columns
beer_permit.drop(columns=['Business Owner', 'Census Tract', 'Council District'], inplace=True)
beer_permit.head()
```

```python
#Drop null values
beer_permit = beer_permit.dropna()
beer_permit
```

```python
#Drop rows with permit subtype OFFSALES
permit_offsales = beer_permit[beer_permit['Permit SubType']=="OFFSALES"].index
beer_permit = beer_permit.drop(permit_offsales)
```
