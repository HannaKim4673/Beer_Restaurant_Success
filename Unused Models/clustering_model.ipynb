```python
import pandas as pd
import hvplot.pandas

import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
df2 = pd.read_csv("database_join.csv")
df2.head()
```

```python

df2.drop(columns=['index'], inplace=True)
df2.head()
```

```python
# Create a new DataFrame that holds only the Permit Number.
permit_number_df = pd.DataFrame(df2["permit_number"])
permit_number_df.head()
```

```python
df2.columns.tolist()
```

```python
permits_by_income_df=df2[["zip_code", "average_household_income","permit_number"]]
permits_by_income_df.head()
```

```python
# Use get_dummies() to create variables for text features.
X = pd.get_dummies(data=permits_by_income_df, columns=["zip_code","permit_number"])
X.head()
```

```python
# Standardize the data with StandardScaler().
permit_scaled = StandardScaler().fit_transform(X)
permit_scaled
```

```python
# Using PCA to reduce dimension to three principal components.
pca = PCA(n_components=3)
permit_pca = pca.fit_transform(permit_scaled)
permit_pca
```

```python
# Create a DataFrame with the three principal components.
pcs_df = pd.DataFrame(data=permit_pca, columns=["pc1", "pc2", "pc3"]
).set_index(permits_by_income_df.index)
pcs_df.head()
```

```python
# Create an elbow curve to find the best value for K.
# Create inertia variable set
inertia =[]
# create list for K values
k = list(range(1, 15))
# loop through K values to determine inertia
for i in k:
    km = KMeans(n_clusters =i, random_state = 42)
    km.fit(pcs_df)
    inertia.append(km.inertia_)

# Create elbow curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x='k', y='inertia', xticks=k, title='Elbow Curve')
```

```python
# Initialize the K-Means model.
model = KMeans(n_clusters=5, random_state=42)

# Fit the model
model.fit(pcs_df)

# Predict clusters
predictions = model.predict(pcs_df)
predictions
```

```python
# Create a new DataFrame including predicted clusters and permit features.
# Concatentate the permit_df and pcs_df DataFrames on the same columns.
clustered_df = pd.concat([permits_by_income_df, pcs_df], axis=1)
clustered_df.head()
```

```python
#  Add back "permit_number". 
clustered_df["permit_number"] = permit_number_df["permit_number"]
clustered_df.head()
```

```python
#  Add a new column, "Class" to the clustered_df DataFrame that holds the predictions.
clustered_df["Class"] = model.labels_
```

```python
# Print the shape of the clustered_df
print(clustered_df.shape)
clustered_df.head(10)
```

```python
# 2D scatter with pca data and clusters
clustered_df.hvplot.scatter(x="pc1", y = "pc2", by="Class")
```

```python
# 3D scatter
fig = px.scatter_3d(clustered_df, x="pc1", y="pc2", z="pc3", color="Class", symbol="Class", width=800, hover_name="zip_code", hover_data=["permit_number"])
fig.update_layout(legend=dict(x=0, y=1))
fig.show()
```

```python
fig = px.scatter_3d(clustered_df, x="zip_code", y="average_household_income", z="permit_number", color="Class", symbol="Class", width=800, hover_name="zip_code", hover_data=["permit_number"])
fig.update_layout(legend=dict(x=0, y=1))
fig.show()
```
